# Top 10 Items Sold

```python
# Get insights for top sales
```

```python
top_sold_items = df.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(10)
top_sold_items # Shows me the top 10 items sold in the list
```

```python
plt.title('Top 10 Items Sold')
top_sold_items.plot(kind='barh', color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Total Sold')

plt.show()
```

# Employee Sales

```python
unique_names = df['Employee'].unique() 
unique_names #Many names in this file, so I condenced the list to see how many names there are
```

```python
top_sales_people = df.groupby('Employee')['OrderTotal'].sum().sort_values(ascending=False).head(3)
top_sales_people
```

```python
margaret_sales = df[df['Employee'] == 'Margaret Peacock']

margaret_top_sell_item = margaret_sales.groupby('ProductName')['Quantity'].sum().sort_values(ascending=False).head(3) #This shows the item that is sold most by Nancy

margaret_top_sell_item
```
